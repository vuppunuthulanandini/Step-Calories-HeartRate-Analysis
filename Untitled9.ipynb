```python
import pandas as pd

# Create a synthetic fitness dataset
data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05'],
    'Steps Walked': [8000, 9000, 12000, 11000, 10000],
    'Calories Burned': [300, 320, 400, 350, 330],
    'Hours Slept': [7, 6.5, 8, 7.5, 7],
    'Average Heart Rate': [75, 78, 72, 74, 76]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Display the first few rows of the dataset
df.head(5)
```

```python
import matplotlib.pyplot as plt

# Line Plot for Steps Walked
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Steps Walked'], marker='o', color='b', label='Steps Walked')
plt.title('Steps Walked Over Time')
plt.xlabel('Date')
plt.ylabel('Steps Walked')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Scatter Plot for Calories Burned vs Steps Walked
plt.figure(figsize=(10, 6))
plt.scatter(df['Steps Walked'], df['Calories Burned'], color='r')
plt.title('Calories Burned vs Steps Walked')
plt.xlabel('Steps Walked')
plt.ylabel('Calories Burned')
plt.grid(True)
plt.show()
```

```python
# Line Plot for Heart Rate with Threshold Line
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Average Heart Rate'], marker='o', color='g', label='Heart Rate')
plt.axhline(y=80, color='r', linestyle='--', label='Threshold (80 bpm)')  # Threshold line
plt.title('Heart Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Average Heart Rate')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Bar Chart for Steps Walked and Calories Burned
width = 0.35  # Bar width
fig, ax = plt.subplots(figsize=(10, 6))

ax.bar(df['Date'], df['Steps Walked'], width, label='Steps Walked')
ax.bar(df['Date'], df['Calories Burned'], width, bottom=df['Steps Walked'], label='Calories Burned')

ax.set_xlabel('Date')
ax.set_ylabel('Total Steps + Calories')
ax.set_title('Steps Walked vs Calories Burned')
ax.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# Stack Plot for Activity Breakdown (Steps Walked, Calories Burned, Hours Slept)
plt.figure(figsize=(10, 6))
plt.stackplot(df['Date'], df['Steps Walked'], df['Calories Burned'], df['Hours Slept'],
              labels=['Steps Walked', 'Calories Burned', 'Hours Slept'], alpha=0.6)
plt.title('Fitness Activity Breakdown Over Time')
plt.xlabel('Date')
plt.ylabel('Values')
plt.legend(loc='upper left')
plt.xticks(rotation=45)
plt.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Interactive Line Plot for Steps Walked
fig = px.line(df, x='Date', y='Steps Walked', title='Steps Walked Over Time', markers=True)
fig.update_layout(xaxis_title='Date', yaxis_title='Steps Walked')
fig.show()
```

```python
# Interactive Box Plot for Hours Slept
fig = px.box(df, y='Hours Slept', title='Distribution of Hours Slept')
fig.update_layout(yaxis_title='Hours Slept')
fig.show()
```

```python
# Interactive Stack Plot for Activity Breakdown
fig = go.Figure()

fig.add_trace(go.Scatter(x=df['Date'], y=df['Steps Walked'], fill='tonexty', mode='lines', name='Steps Walked'))
fig.add_trace(go.Scatter(x=df['Date'], y=df['Calories Burned'], fill='tonexty', mode='lines', name='Calories Burned'))
fig.add_trace(go.Scatter(x=df['Date'], y=df['Hours Slept'], fill='tonexty', mode='lines', name='Hours Slept'))

fig.update_layout(title='Fitness Activity Breakdown Over Time', xaxis_title='Date', yaxis_title='Values')
fig.show()
```
